```python
import pandas as pd
df = pd.read_csv('../datasets/Sleep_Efficiency.csv')
```

### Histograms (Numerical Variables)
* Use seaborn.histplot() to explore distribution.

**Recommended columns:**
* Sleep duration
* Sleep efficiency
* REM sleep percentage
* Caffeine consumption

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Sleep duration'], kde=True)
plt.title('Distribution of Sleep Duration')
plt.xlabel('Hours')
plt.ylabel('People')
plt.tight_layout()
plt.show() 
```

### Box Plots (Numerical Variables)
* Explore spread, skew, and outliers.

**Recommended columns:**
* Sleep efficiency
* Exercise frequency
* Alcohol consumption

```python
sns.boxplot(y=df['Sleep efficiency'])
plt.title('Box Plot of Sleep Efficiency')
plt.ylabel('Efficiency (%)')
plt.tight_layout()
plt.show()
```

### Bar Charts (Categorical Frequencies)
* Use sns.countplot() for category counts.

**Recommended columns:**
* Gender
* Smoking Status
* BMI category

```python
sns.countplot(x='Smoking status', data=df)
plt.title('Smoking status Distribution')
plt.xlabel('Status')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
