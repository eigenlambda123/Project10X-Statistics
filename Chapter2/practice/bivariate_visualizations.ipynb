```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('../datasets/StudentsPerformance.csv')
```

### 1. Scatter Plot with Regression Line
* There is a strong positive correlation between Math and Reading scores â€” students who perform well in math tend to also perform well in reading.

```python
sns.lmplot(x='math score', y='reading score', data=df, aspect=1.5)
plt.title('Scatter Plot with Regression Line: Math Score vs Reading Score')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.tight_layout()
plt.show()
```

### 3. Scatter plot with hue
* The relationship between Math and Writing scores holds across genders, with slight variations in score distributions.

```python
sns.lmplot(x='math score', y='writing score', hue='gender', data=df, aspect=1.5)
plt.title('Math Score vs Writing Score by Gender')
plt.xlabel('Math Score')
plt.ylabel('Writing Score')
plt.tight_layout()
plt.show()
```

### 4. Grouped box plots
* Students with standard lunch tend to have higher median writing scores compared to those with free/reduced lunch.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='lunch', y='writing score', data=df)
plt.title('Writing Score Distribution by Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('Writing Score')
plt.tight_layout()
plt.show()
```
