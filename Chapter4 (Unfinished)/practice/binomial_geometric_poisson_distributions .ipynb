### **Computations**

```python
from scipy.stats import binom, geom, poisson

# Parameters
n = 10  # number of sample
p = 0.4 # probability
λ = 3   # mean
k = 3   # desired outcome

# Binomial: P(X = 3)
binom_prob = binom.pmf(k, n, p)

# Geometric: P(X = 3)
geom_prob = geom.pmf(k, p)

# Poisson: P(X = 3)
poisson_prob = poisson.pmf(k, λ)

print(f"Binomial P(X=3): {binom_prob:.4f}")
print(f"Geometric P(X=3): {geom_prob:.4f}")
print(f"Poisson P(X=3): {poisson_prob:.4f}")
```

### **Distributions**

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(1, 3, figsize=(18, 4))

# BINOMIAL
x_binom = np.arange(0, n + 1)
axs[0].bar(x_binom, binom.pmf(x_binom, n, p), color='skyblue')
axs[0].set_title('Binomial PMF (n=10, p=0.4)')
axs[0].set_xlabel('k')
axs[0].set_ylabel('P(X=k)')

# GEOMETRIC
x_geom = np.arange(1, 11)
axs[1].bar(x_geom, geom.pmf(x_geom, p), color='salmon')
axs[1].set_title('Geometric PMF (p=0.4)')
axs[1].set_xlabel('k')
axs[1].set_ylabel('P(X=k)')

# POISSON
x_poisson = np.arange(0, 11)
axs[2].bar(x_poisson, poisson.pmf(x_poisson, λ), color='lightgreen')
axs[2].set_title('Poisson PMF (λ=3)')
axs[2].set_xlabel('k')
axs[2].set_ylabel('P(X=k)')

plt.tight_layout()
plt.show()
```
