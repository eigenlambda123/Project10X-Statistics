### **Problem 1**

A factory produces two types of products: **Type A** and **Type B**.
From records:

* $70\%$ of products are **Type A**
* $30\%$ of products are **Type B**
* $5\%$ of Type A products are defective
* $8\%$ of Type B products are defective

**Find:**

1. $P(\text{Defective} \mid \text{Type A})$ — probability a product is defective given it is Type A.
2. $P(\text{Type A} \mid \text{Defective})$ — probability a product is Type A given it is defective.


### **Solution 1**

**Given:**

* $P(\text{Type A}) = 0.70$
* $P(\text{Type B}) = 0.30$
* $P(\text{Defective} \mid \text{Type A}) = 0.05$
* $P(\text{Defective} \mid \text{Type B}) = 0.08$


**1. $P(\text{Defective} \mid \text{Type A})$**

This is given directly:

$$
P(\text{Defective} \mid \text{Type A}) = 0.05.
$$


**2. $P(\text{Type A} \mid \text{Defective})$**

Use Bayes’ Rule:

$$
P(\text{Type A} \mid \text{Defective})
= \frac{P(\text{Defective} \mid \text{Type A}) \cdot P(\text{Type A})}{P(\text{Defective})}.
$$

First compute $P(\text{Defective})$ using the law of total probability:

$$
\begin{aligned}
P(\text{Defective})
&= P(\text{Defective} \mid \text{Type A})P(\text{Type A})
  + P(\text{Defective} \mid \text{Type B})P(\text{Type B}) \\
&= (0.05)(0.70) + (0.08)(0.30) \\
&= 0.035 + 0.024 = 0.059.
\end{aligned}
$$

Now apply Bayes’ Rule:

$$
\begin{aligned}
P(\text{Type A} \mid \text{Defective})
&= \frac{0.05 \times 0.70}{0.059}
= \frac{0.035}{0.059}
= \frac{35}{59} \\
&\approx 0.5932 \approx 59.32\%.
\end{aligned}
$$

**Final answers**

* $P(\text{Defective} \mid \text{Type A}) = 0.05$
* $P(\text{Type A} \mid \text{Defective}) = \dfrac{35}{59} \approx 0.5932 \;(\approx 59.32\%)$

### **Problem 2**

Out of **100 days**:

* It rained on **30** days.
* On **20** of those rainy days, the forecast had predicted rain.
* On **10** of the non-rainy days, the forecast also predicted rain.

Find:

1. $P(\text{Predicted Rain} \mid \text{Rain})$
2. $P(\text{Rain} \mid \text{Predicted Rain})$

### **Solution 2**

**Step 1 — Organize the data into a table**

|             | Predicted Rain | Predicted No Rain | Total |
| ----------- | -------------- | ----------------- | ----- |
| **Rain**    | 20             | 10                | 30    |
| **No Rain** | 10             | 60                | 70    |
| **Total**   | 30             | 70                | 100   |


**a. $P(\text{Predicted Rain} \mid \text{Rain})$**

$$
P(\text{Predicted Rain} \mid \text{Rain}) = \frac{\text{Rain AND Predicted Rain}}{\text{Rain}} = \frac{20}{30} = \frac{2}{3} \approx 0.6667
$$


**b. $P(\text{Rain} \mid \text{Predicted Rain})$**

$$
P(\text{Rain} \mid \text{Predicted Rain}) = \frac{\text{Rain AND Predicted Rain}}{\text{Predicted Rain}} = \frac{20}{30} = \frac{2}{3} \approx 0.6667
$$

**Interpretation**:
Both probabilities happen to be the same here (about **66.67%**), but that’s **not always the case** in other datasets. It means that when it rains, the forecast is correct two-thirds of the time, and when the forecast says rain, it’s correct two-thirds of the time.

### **Problem 3:**

In a school of 200 students:

* 120 students study **Spanish**.
* 80 students study **French**.
* 30 students study **both** Spanish and French.

Find:

1. $P(\text{French} \mid \text{Spanish})$ — probability a student studies French given they study Spanish.
2. $P(\text{Spanish} \mid \text{French})$ — probability a student studies Spanish given they study French.


### **Solution 3**

> In a school of 200 students:

> * 120 students study **Spanish**.
> * 80 students study **French**.
> * 30 students study **both** Spanish and French.

**1.** $P(\text{French} \mid \text{Spanish})$ — probability a student studies French given they study Spanish.

$$
P(\text{French}\mid\text{Spanish})
= \frac{P(\text{French}\cap\text{Spanish})}{P(\text{Spanish})}
= \frac{30/200}{120/200}
= \frac{30}{120}
= \frac{1}{4}
= 0.25
$$

**2.** $P(\text{Spanish} \mid \text{French})$ — probability a student studies Spanish given they study French.

$$
P(\text{Spanish}\mid\text{French})
= \frac{P(\text{Spanish}\cap\text{French})}{P(\text{French})}
= \frac{30/200}{80/200}
= \frac{30}{80}
= \frac{3}{8}
= 0.375
$$

### **Problem 4**

A company tests 500 gadgets. The results:

|             | Pass Test | Fail Test | Total |
| ----------- | --------- | --------- | ----- |
| **Model X** | 180       | 20        | 200   |
| **Model Y** | 250       | 50        | 300   |
| **Total**   | 430       | 70        | 500   |

Find:

1. $P(\text{Fail} \mid \text{Model Y})$ — probability a gadget fails given it’s Model Y.
2. $P(\text{Model Y} \mid \text{Fail})$ — probability the gadget is Model Y given it failed.



### **Solution 4**

> Test results for 500 gadgets:

|             | Pass Test | Fail Test | Total |
| ----------- | --------- | --------- | ----- |
| **Model X** | 180       | 20        | 200   |
| **Model Y** | 250       | 50        | 300   |
| **Total**   | 430       | 70        | 500   |

**1.** $P(\text{Fail} \mid \text{Model Y})$

$$
P(\text{Fail}\mid\text{Model Y})
= \frac{\text{Fail \& Model Y}}{\text{Total Model Y}}
= \frac{50}{300}
= \frac{1}{6}
\approx 0.1667
$$

**2.** $P(\text{Model Y} \mid \text{Fail})$

$$
P(\text{Model Y}\mid\text{Fail})
= \frac{\text{Fail \& Model Y}}{\text{Total Fail}}
= \frac{50}{70}
= \frac{5}{7}
\approx 0.7143
$$
