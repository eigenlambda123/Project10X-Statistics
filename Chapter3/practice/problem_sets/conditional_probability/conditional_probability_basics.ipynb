### **Problem 1**

A card is drawn from a standard deck of 52 cards.

Let:

* $A$: “The card is a King”
* $B$: “The card is a Spade”

Find:

1. $P(A \mid B)$ — probability the card is a King given it’s a Spade.
2. $P(B \mid A)$ — probability the card is a Spade given it’s a King.


### **Solution 1**

Let:

* $A$: “The card is a King”
* $B$: “The card is a Spade”


**Step 1:** Find $P(A \cap B)$

* Only 1 King of Spades in the deck:

$$
P(A \cap B) = \frac{1}{52}
$$

**Step 2:** Find $P(B)$ and $P(A)$

* $P(B) = \frac{13}{52} = \frac{1}{4}$ (13 spades)
* $P(A) = \frac{4}{52} = \frac{1}{13}$ (4 kings)

**(a) $P(A \mid B)$:**

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)} = \frac{\frac{1}{52}}{\frac{1}{4}} = \frac{1}{13}
$$

**(b) $P(B \mid A)$:**

$$
P(B \mid A) = \frac{P(A \cap B)}{P(A)} = \frac{\frac{1}{52}}{\frac{1}{13}} = \frac{1}{4}
$$


**Final Answers:**

* $P(A \mid B) = \frac{1}{13}$
* $P(B \mid A) = \frac{1}{4}$

### **Problem 2**

In a school:

* $60\%$ of students play basketball.
* $25\%$ of these also play soccer.
* Overall, $40\%$ of all students play soccer.

Find:

1. $P(\text{Soccer} \mid \text{Basketball})$
2. $P(\text{Basketball} \mid \text{Soccer})$

### **Solution 2**

Let:

* $B$: plays basketball
* $S$: plays soccer


**Given:**

* $P(B) = 0.60$
* $P(S \mid B) = 0.25$
* $P(S) = 0.40$


**(a) $P(S \mid B)$:**
Already given:

$$
P(S \mid B) = 0.25
$$

**(b) $P(B \mid S)$:**
First find $P(S \cap B)$:

$$
P(S \cap B) = P(S \mid B) \cdot P(B) = 0.25 \times 0.60 = 0.15
$$

Then:

$$
P(B \mid S) = \frac{P(S \cap B)}{P(S)} = \frac{0.15}{0.40} = 0.375
$$

**Final Answers:**

* $P(S \mid B) = 0.25$
* $P(B \mid S) = 0.375$
