### **Problem 1**

Two fair dice are rolled.  

Let:  
- **A**: “Sum is greater than 8”  
- **B**: “At least one die is a 5”  

Find:  

1. $P(A \mid B)$ — probability the sum is greater than 8 given at least one die is a 5.  
2. $P(B \mid A)$ — probability at least one die is a 5 given the sum is greater than 8.  

### **Solution 1**

Two fair dice → sample space size \$|S| = 6 \times 6 = 36\$ (ordered pairs \$(i,j)\$).

**Define the events**

* \$A = {\text{sum is greater than }8} = {\text{sum }=9,10,11,12}\$
  List outcomes:

  $$
  A=\{(3,6),(4,5),(5,4),(6,3),\ (4,6),(5,5),(6,4),\ (5,6),(6,5),\ (6,6)\}.
  $$

  So \$|A| = 10\$.

* \$B = {\text{at least one die is }5}\$
  List outcomes:

  $$
  B=\{(5,1),(5,2),(5,3),(5,4),(5,5),(5,6),\ (1,5),(2,5),(3,5),(4,5),(6,5)\}.
  $$

  So \$|B| = 11\$.

* Intersection \$A\cap B\$ (sum \$>8\$ **and** at least one 5):

  $$
  A\cap B=\{(4,5),(5,4),(5,5),(5,6),(6,5)\}.
  $$

  So \$|A\cap B| = 5\$.


**1. Compute \$P(A \mid B)\$**

By definition,

$$
P(A\mid B)=\frac{P(A\cap B)}{P(B)}
= \frac{\frac{|A\cap B|}{36}}{\frac{|B|}{36}}
= \frac{|A\cap B|}{|B|}
= \frac{5}{11}.
$$

Numerical value: \$P(A\mid B)=\dfrac{5}{11}\approx 0.4545\$.

**2. Compute \$P(B \mid A)\$**

By definition,

$$
P(B\mid A)=\frac{P(A\cap B)}{P(A)}
= \frac{\frac{|A\cap B|}{36}}{\frac{|A|}{36}}
= \frac{|A\cap B|}{|A|}
= \frac{5}{10}=\frac{1}{2}.
$$

Numerical value: \$P(B\mid A)=\tfrac{1}{2}=0.5\$.

### **Final answers**

* \$P(A\mid B)=\dfrac{5}{11}\approx 0.4545\$
* \$P(B\mid A)=\dfrac{1}{2}=0.5\$

### **Problem 2**

A standard deck of \$52\$ cards is used. Two cards are drawn **without replacement**.

Let:

* \$A\$: "The first card is a King"
* \$B\$: "The second card is a Heart"

Find:

1. \$P(B \mid A)\$ — probability the second card is a Heart given the first card is a King.
2. \$P(A \mid B)\$ — probability the first card is a King given the second card is a Heart.




### **Solution 2**

**Setup:** Two cards are drawn **without replacement** from a standard 52-card deck.

* $A$: first card is a King.
* $B$: second card is a Heart.
  Total ordered outcomes = $52 \times 51$.

**1. $P(B \mid A)$ — probability the second card is a Heart given the first card is a King**

Condition on which King was drawn first:

* If the first card is the King of Hearts (1 of the 4 kings), then remaining hearts = 12, so probability second is heart = $\dfrac{12}{51}$.
* If the first card is a non-heart King (3 of the 4 kings), then remaining hearts = 13, so probability second is heart = $\dfrac{13}{51}$.

Thus

$$
\begin{aligned}
P(B\mid A)
&= \frac{1}{4}\cdot\frac{12}{51} \;+\; \frac{3}{4}\cdot\frac{13}{51} \\
&= \frac{12 + 39}{4\cdot 51}
= \frac{51}{204}
= \frac{1}{4}.
\end{aligned}
$$

So

$$
\boxed{\,P(B\mid A)=\tfrac{1}{4}=0.25\,}
$$

(Reason: by symmetry the probability a randomly chosen card is a heart is $13/52=1/4$; knowing the first is a King does not change the chance that the second is a heart.)

**2. $P(A \mid B)$ — probability the first card is a King given the second card is a Heart**

Compute $P(A\cap B)$ first. Count ordered favorable pairs where first is a King and second is a Heart:

* If first = King of Hearts: remaining hearts = 12 → $1\times 12 = 12$ ordered pairs.
* If first = non-heart King (3 choices): remaining hearts = 13 each → $3\times 13 = 39$ ordered pairs.

Total favorable ordered pairs = $12 + 39 = 51$. So

$$
P(A\cap B)=\frac{51}{52\cdot 51}=\frac{1}{52}.
$$

Also $P(B)$ (second card is a Heart) by symmetry is

$$
P(B)=\frac{13}{52}=\frac{1}{4}.
$$

Now apply conditional probability:

$$
P(A\mid B)=\frac{P(A\cap B)}{P(B)}
= \frac{\tfrac{1}{52}}{\tfrac{1}{4}}
= \frac{4}{52}
= \frac{1}{13}.
$$

So

$$
\boxed{\,P(A\mid B)=\tfrac{1}{13}\approx 0.07692\,}
$$

(Again by symmetry this equals the unconditional probability that the first card is a King, $4/52=1/13$.)

### **Final answers**

* $P(B\mid A)=\dfrac{1}{4}=0.25$
* $P(A\mid B)=\dfrac{1}{13}\approx 0.07692$

### **Problem 3**

A fair coin is flipped three times.

Let:

* \$A\$: "At least two heads occur"
* \$B\$: "The first flip is heads"

Find:

1. \$P(A \mid B)\$
2. \$P(B \mid A)\$



### **Solution 3**

**Step 1 — List the sample space**

A fair coin flipped three times has $2^3 = 8$ equally likely outcomes:

$$
\{ HHH, HHT, HTH, HTT, THH, THT, TTH, TTT \}
$$

**1. Define the events**

* **Event $A$**: "At least two heads"
  Outcomes: $\{ HHH, HHT, HTH, THH \}$ (3 heads or exactly 2 heads)
  Count: $4$ outcomes.

* **Event $B$**: "First flip is heads"
  Outcomes: $\{ HHH, HHT, HTH, HTT \}$
  Count: $4$ outcomes.

**3. Intersection $A \cap B$**

Outcomes that satisfy **at least two heads** **and** **first is heads**:

$$
A \cap B = \{ HHH, HHT, HTH \}
$$

Count: $3$ outcomes.

**4. $P(A \mid B)$**

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}
= \frac{\frac{3}{8}}{\frac{4}{8}}
= \frac{3}{4}
= 0.75
$$

**5. $P(B \mid A)$**

$$
P(B \mid A) = \frac{P(A \cap B)}{P(A)}
= \frac{\frac{3}{8}}{\frac{4}{8}}
= \frac{3}{4}
= 0.75
$$

### **Final Answers**

* $P(A \mid B) = \frac{3}{4} = 0.75$
* $P(B \mid A) = \frac{3}{4} = 0.75$

### **Problem 4**

A bag contains \$5\$ red balls and \$7\$ blue balls. Two balls are drawn **without replacement**.

Let:

* \$A\$: "Both balls are the same color"
* \$B\$: "The first ball is red"

Find:

1. \$P(A \mid B)\$
2. \$P(B \mid A)\$


### **Solution 4**

#### **Total Balls**

* Red: $5$
* Blue: $7$
* Total: $12$


#### **$P(A \mid B)$**

If the first ball is red, remaining balls: $4$ red, $7$ blue (total $11$).

$$
P(A \mid B) = \frac{4}{11}
$$


#### **$P(B \mid A)$**

$$
P(A) = \frac{5}{12} \cdot \frac{4}{11} + \frac{7}{12} \cdot \frac{6}{11}
= \frac{20}{132} + \frac{42}{132} = \frac{31}{66}
$$

$$
P(A \cap B) = \frac{5}{12} \cdot \frac{4}{11} = \frac{10}{66}
$$

$$
P(B \mid A) = \frac{\frac{10}{66}}{\frac{31}{66}} = \frac{10}{31}
$$

### **Final Answers**

* $P(A \mid B) = \frac{4}{11} \approx 0.3636$
* $P(B \mid A) = \frac{10}{31} \approx 0.3226$
