### **Problem 1**

> A card is drawn from a standard deck. Define:

- The sample space
- Event $A$ = “drawing a face card”
- Event $B$ = “drawing a red card”
- $A \cap B$, $A \cup B$, $A^c$


### **Solutions 1**

**1. Define the Sample Space**

> The **sample space** $S$ is the set of all 52 cards in a standard deck:

$$
S = \{ \text{Ace, 2, ..., 10, Jack, Queen, King of Hearts, Diamonds, Clubs, Spades} \}
$$

**2. Event $A$: “drawing a face card”**

> Face cards are **Jack, Queen, King**.
There are 3 face cards per suit × 4 suits = 12 total.

$$
A = \{\text{J♥, Q♥, K♥, J♦, Q♦, K♦, J♣, Q♣, K♣, J♠, Q♠, K♠}\}
$$

**3. Event $B$: “drawing a red card”**

> Red cards = Hearts and Diamonds = 13 + 13 = 26 cards.

$$
B = \{ \text{All 13 cards of ♥ and 13 cards of ♦} \}
$$

**4. Find $A \cap B$: “face cards that are red”**

> Intersection = face cards **that are red (♥ or ♦)**:

$$
A \cap B = \{\text{J♥, Q♥, K♥, J♦, Q♦, K♦}\}
$$

> Total of **6 cards**.

**5. Find $A \cup B$: “either a face card or a red card or both”**

> Union = all red cards **plus** the **black** face cards (♠ and ♣):

* Red cards = 26
* Black face cards = J♠, Q♠, K♠, J♣, Q♣, K♣ → 6 cards

$$
A \cup B = B \cup \{\text{J♠, Q♠, K♠, J♣, Q♣, K♣}\}
$$

So:

$$
|A \cup B| = 26 + 6 = 32 \text{ cards}
$$

**6. Find $A^c$: “not a face card”**

> Complement of $A$ = all cards that are **not** Jack, Queen, or King.

Total cards = 52
Face cards = 12
So:

$$
A^c = 52 - 12 = 40 \text{ cards}
$$

Or:

$$
A^c = \{\text{All Aces and number cards (2–10) from all suits}\}
$$

### **Problem 2**

> You roll two dice. Let event \$C\$ = “sum is 7”, event \$D\$ = “at least one die is a 4”.

* List all outcomes in \$C\$ and \$D\$
* What is \$C \cap D\$?

### **Solutions 2**

**1. Define the Sample Space**

> When rolling two dice, the **sample space** \$S\$ consists of all ordered pairs:

$$
S = \{(i, j) \mid i, j \in \{1, 2, 3, 4, 5, 6\} \}
$$

There are \$6 \times 6 = 36\$ total outcomes.


**2. Event \$C\$: “sum is 7”**

> The outcomes where the **sum of the two dice is 7**:

$$
C = \{(1,6), (2,5), (3,4), (4,3), (5,2), (6,1)\}
$$

There are **6 outcomes** in \$C\$.


**3. Event \$D\$: “at least one die is a 4”**

> Outcomes where **at least one die is a 4**:

* First die is 4:
  \$(4,1), (4,2), (4,3), (4,4), (4,5), (4,6)\$
* Second die is 4:
  \$(1,4), (2,4), (3,4), (5,4), (6,4)\$

(Note: \$(4,4)\$ is already included above)

So:

$$
D = \{(4,1), (4,2), (4,3), (4,4), (4,5), (4,6), (1,4), (2,4), (3,4), (5,4), (6,4)\}
$$

There are **11 outcomes** in \$D\$.


**4. Find \$C \cap D\$: “sum is 7 and at least one die is a 4”**

We take the **intersection** of sets \$C\$ and \$D\$:

From \$C\$:
\${(1,6), (2,5), (3,4), (4,3), (5,2), (6,1)}\$

From \$D\$:
\${(4,1), (4,2), (4,3), (4,4), (4,5), (4,6), (1,4), (2,4), (3,4), (5,4), (6,4)}\$

Intersection:

* \$(3,4)\$
* \$(4,3)\$

So:

$$
C \cap D = \{(3,4), (4,3)\}
$$

There are **2 outcomes** in \$C \cap D\$.
