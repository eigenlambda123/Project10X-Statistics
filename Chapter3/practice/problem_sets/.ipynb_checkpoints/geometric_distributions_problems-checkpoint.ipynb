### **Problem 1**
Andrea practices shots at goal. For each shot, the probability of scoring a goal is $\frac{2}{5}$. Each shot is independent of other shots.

#### Find the probability that she scores her first goal:
**(a)** On her 5th shot. [2 marks]  
**(b)** Before her 5th shot. [3 marks]

**Given:**
- Probability of success per shot: $p = \frac{2}{5}$
- Shots are independent
- Geometric distribution applies

### **Solution 1**

#### **(a) Probability that she scores her first goal on her 5th shot**
Using the Geometric Distribution formula:
$$
P(X = x) = (1 - p)^{x - 1} \cdot p
$$

**Substitute \( x = 5 \) and \( p = 0.4 \):**
$$
P(X = 5) = (1 - 0.4)^{5 - 1} \cdot 0.4 = (0.6)^4 \cdot 0.4
$$

**Calculate:**
$$
P(X = 5) = 0.1296 \cdot 0.4 = 0.05184
$$

**Convert to percentage:**
$$
0.05184 \times 100\% = 5.184\%
$$

**Final Answer:**
$$
\boxed{5.18\%} \quad \text{(rounded to 2 decimal places)}
$$


#### **(b) Probability that she scores her first goal before her 5th shot**
Using the Cumulative Geometric Distribution formula:
$$
P(X \leq k) = 1 - (1 - p)^k
$$

**Substitute \( k = 4 \) and \( p = 0.4 \):**
$$
P(X \leq 4) = 1 - (1 - 0.4)^4 = 1 - (0.6)^4
$$

**Calculate:**
$$
P(X \leq 4) = 1 - 0.1296 = 0.8704
$$

**Convert to percentage:**
$$
0.8704 \times 100\% = 87.04\%
$$

**Final Answer:**
$$
\boxed{87.04\%} \quad \text{(rounded to 2 decimal places)}
$$

### **Problem 2**  
*A biased coin has a probability of $\frac{2}{3}$ to land heads on any throw. The coin is thrown repeatedly.*

Let $X$ be the number of throws **up to and including the first head**.

**Find:**  
**(a)** $P(X = 4)$.  * 

**(b)** $P(X < 4)$.  

**(c)** $E(X)$.  


**Given:**  
- Probability of heads: $p = \frac{2}{3}$  
- $X$ follows a **geometric distribution** (counts trials until first success).  


### **Solution 2**  

#### **(a) $P(X = 4)$**  
Using the geometric distribution formula for the first success on the $x$-th trial:  
$$
P(X = x) = (1 - p)^{x - 1} \cdot p
$$  
Substitute $x = 4$ and $p = \frac{2}{3}$:  
$$
P(X = 4) = \left(1 - \frac{2}{3}\right)^{3} \cdot \frac{2}{3} = \left(\frac{1}{3}\right)^3 \cdot \frac{2}{3} = \frac{2}{81}
$$  
Convert to percentage:  
$$
\frac{2}{81} \times 100\% \approx 2.47\%
$$  
**Final Answer:**  
$$
\boxed{2.47\%}
$$

#### **(b) $P(X < 4)$**  
Since $X < 4$ includes trials $1$, $2$, and $3$, we use the cumulative probability formula:  
$$
P(X \leq 3) = 1 - (1 - p)^3
$$  
Substitute $p = \frac{2}{3}$:  
$$
P(X \leq 3) = 1 - \left(\frac{1}{3}\right)^3 = 1 - \frac{1}{27} = \frac{26}{27}
$$  
Convert to percentage:  
$$
\frac{26}{27} \times 100\% \approx 96.30\%
$$  
**Final Answer:**  
$$
\boxed{96.30\%}
$$

#### **(c) $E(X)$**  
For a geometric distribution, the expected value is:  
$$
E(X) = \frac{1}{p} = \frac{3}{2} = 1.5 \text{ throws}
$$  
**Final Answer:**  
$$
\boxed{1.5}
$$

### **Problem 3**
*30% of people own a Talk-2 phone. Individuals are selected at random, one at a time, and asked about ownership. The number of people questioned up to and including the first Talk-2 phone owner is denoted by $X$.*

**Find:**  
**(a)** $P(X = 4)$.   

**(b)** $P(X > 4)$.  

**(c)** $P(X < 6)$. 

**Given:**  
- Probability of owning a Talk-2 phone: $p = 0.3$  
- $X$ follows a **geometric distribution** (trials until first success). 

### **Solution 3**

#### **(a) $P(X = 4)$**  
*Probability the first Talk-2 owner is the 4th person questioned.*

Using the geometric PMF:
$$
P(X = k) = (1 - p)^{k-1} \cdot p
$$

Substitute $k = 4$, $p = 0.3$:
$$
P(X = 4) = (0.7)^3 \times 0.3 = 0.343 \times 0.3 = 0.1029
$$

**Final Answer:**
$$
\boxed{10.29\%}
$$


#### **(b) $P(X > 4)$**  
*Probability it takes more than 4 people to find the first Talk-2 owner.*

Using the geometric survival function:
$$
P(X > k) = (1 - p)^k
$$

Substitute $k = 4$, $p = 0.3$:
$$
P(X > 4) = (0.7)^4 = 0.2401
$$

**Final Answer:**
$$
\boxed{24.01\%}
$$


#### **(c) $P(X < 6)$**  
*Probability the first Talk-2 owner is found within the first 5 people.*

Using the geometric CDF:
$$
P(X \leq k) = 1 - (1 - p)^k
$$

Substitute $k = 5$, $p = 0.3$:
$$
P(X < 6) = P(X \leq 5) = 1 - (0.7)^5 = 1 - 0.16807 = 0.83193
$$

**Final Answer:**
$$
\boxed{83.19\%}
$$
