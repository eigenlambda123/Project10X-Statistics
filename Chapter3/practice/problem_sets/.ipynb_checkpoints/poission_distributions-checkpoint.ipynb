### **Problem 1**

A call center receives an average of 3 customer calls per 10 minutes.

#### Find the probability that:

**(a)** Exactly 5 calls arrive in a 10-minute interval.

**(b)** At most 2 calls arrive in the same interval.

**Given:**

* Average rate: \$\lambda = 3\$ calls per 10 minutes
* Poisson distribution applies

### **Solution 1**

#### **(a) Probability that exactly 5 calls arrive in 10 minutes**

This is a **Poisson distribution** problem.
Use the formula:

$$
P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}
$$

**Given:**

* \$\lambda = 3\$
* \$k = 5\$

**Substitute values:**

$$
P(X = 5) = \frac{e^{-3} \cdot 3^5}{5!}
= \frac{e^{-3} \cdot 243}{120}
$$

**Calculate:**

* \$e^{-3} \approx 0.04979\$
* \$\frac{243}{120} = 2.025\$

Then:

$$
P(X = 5) \approx 0.04979 \cdot 2.025 = 0.1008
$$

**Convert to percentage:**

$$
0.1008 \times 100\% = 10.08\%
$$

**Final Answer:**

$$
\boxed{10.08\%} \quad \text{(rounded to 2 decimal places)}
$$


#### **(b) Probability that at most 2 calls arrive**

We want:

$$
P(X \leq 2) = P(0) + P(1) + P(2)
$$

Use the Poisson formula for each:

* \$P(0) = \frac{e^{-3} \cdot 3^0}{0!} = e^{-3} \cdot 1\$
* \$P(1) = \frac{e^{-3} \cdot 3^1}{1!} = e^{-3} \cdot 3\$
* \$P(2) = \frac{e^{-3} \cdot 3^2}{2!} = e^{-3} \cdot \frac{9}{2}\$

**Factor out \$e^{-3}\$:**

$$
P(X \leq 2) = e^{-3} \left(1 + 3 + \frac{9}{2} \right)
= e^{-3} \cdot \left(1 + 3 + 4.5\right) = e^{-3} \cdot 8.5
$$

**Use \$e^{-3} \approx 0.04979\$:**

$$
P(X \leq 2) \approx 0.04979 \cdot 8.5 = 0.4232
$$

**Convert to percentage:**

$$
0.4232 \times 100\% = 42.32\%
$$

**Final Answer:**

$$
\boxed{42.32\%} \quad \text{(rounded to 2 decimal places)}
$$

### **Problem 2**

A hospital records an average of 2 emergency cases arriving per hour.

#### Find the probability that:

**(a)** No emergencies arrive in a given hour. 

**(b)** More than 3 emergencies arrive in an hour.

**Given:**

* \$\lambda = 2\$ per hour
* Use complement rule for part (b): \$P(X > 3) = 1 - P(X \le 3)\$



### **Solution 2**

#### **(a) Probability that no emergencies arrive in 1 hour**

We use the **Poisson distribution**:

$$
P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}
$$

**Given:**

* \$\lambda = 2\$ (avg emergencies/hour)
* \$k = 0\$

**Substitute:**

$$
P(X = 0) = \frac{e^{-2} \cdot 2^0}{0!} = e^{-2} \cdot 1
$$

**Calculate:**

* \$e^{-2} \approx 0.1353\$

**Final Answer:**

$$
\boxed{13.53\%}
$$


#### **(b) Probability that more than 3 emergencies arrive**

We want:

$$
P(X > 3) = 1 - P(X \leq 3) = 1 - [P(0) + P(1) + P(2) + P(3)]
$$

Letâ€™s compute each term:

* \$P(0) = e^{-2} \cdot \frac{2^0}{0!} = e^{-2} \cdot 1 = 0.1353\$
* \$P(1) = e^{-2} \cdot \frac{2^1}{1!} = e^{-2} \cdot 2 = 0.2706\$
* \$P(2) = e^{-2} \cdot \frac{2^2}{2!} = e^{-2} \cdot 2 = 0.2706\$
* \$P(3) = e^{-2} \cdot \frac{2^3}{3!} = e^{-2} \cdot \frac{8}{6} \approx 0.1804\$

**Add them up:**

$$
P(X \leq 3) = 0.1353 + 0.2706 + 0.2706 + 0.1804 = 0.8569
$$

Then:

$$
P(X > 3) = 1 - 0.8569 = 0.1431
$$

**Convert to percentage:**

$$
0.1431 \times 100\% = 14.31\%
$$

**Final Answer:**

$$
\boxed{14.31\%} \quad \text{(rounded to 2 decimal places)}
$$

### **Problem 3**

A small server receives an average of 4 requests per minute.

#### Find the probability that:

**(a)** Exactly 4 requests are received in 1 minute.

**(b)** Fewer than 4 requests are received in a minute.

**Given:**

* \$\lambda = 4\$ per minute
* Use \$P(X < 4) = P(0) + P(1) + P(2) + P(3)\$



### **Solution 3**

#### **(a) Probability that exactly 4 requests are received in 1 minute**

We use the **Poisson distribution**:

$$
P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}
$$

**Given:**

* \$\lambda = 4\$
* \$k = 4\$

**Substitute:**

$$
P(X = 4) = \frac{e^{-4} \cdot 4^4}{4!} = \frac{e^{-4} \cdot 256}{24}
$$

**Calculate:**

* \$e^{-4} \approx 0.0183\$
* \$\frac{256}{24} \approx 10.67\$

Then:

$$
P(X = 4) \approx 0.0183 \cdot 10.67 = 0.1953
$$

**Convert to percentage:**

$$
0.1953 \times 100\% = 19.53\%
$$

**Final Answer:**

$$
\boxed{19.53\%}
$$


#### **(b) Probability that fewer than 4 requests are received**

We want:

$$
P(X < 4) = P(0) + P(1) + P(2) + P(3)
$$

Compute each:

* \$P(0) = e^{-4} \cdot \frac{4^0}{0!} = e^{-4} \cdot 1 = 0.0183\$
* \$P(1) = e^{-4} \cdot \frac{4^1}{1!} = e^{-4} \cdot 4 = 0.0733\$
* \$P(2) = e^{-4} \cdot \frac{4^2}{2!} = e^{-4} \cdot 8 = 0.1465\$
* \$P(3) = e^{-4} \cdot \frac{4^3}{3!} = e^{-4} \cdot \frac{64}{6} \approx 0.1953\$

**Add them up:**

$$
P(X < 4) = 0.0183 + 0.0733 + 0.1465 + 0.1953 = 0.4334
$$

**Convert to percentage:**

$$
0.4334 \times 100\% = 43.34\%
$$

**Final Answer:**

$$
\boxed{43.34\%}
$$
