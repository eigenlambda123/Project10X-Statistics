### **Problem 1**

> A fair six-sided die is rolled.

**a.** What is the probability of rolling a 4?
**b.** What is the probability of rolling an odd number?


### **Solution 1**

Let the sample space be:

$$
S = \{1, 2, 3, 4, 5, 6\}, \quad |S| = 6
$$

**a.** Rolling a 4:

* Favorable outcome: \${4}\$
* Total outcomes: \$6\$

$$
P(4) = \frac{1}{6}
$$

**b.** Rolling an odd number:

* Favorable outcomes: \${1, 3, 5}\$ (there are 3)

$$
P(\text{odd}) = \frac{3}{6} = \frac{1}{2}
$$

### **Problem 2**

> A card is drawn at random from a standard deck of 52 cards.

**a.** What is the probability of drawing a Queen?
**b.** What is the probability of drawing a red card or a King?


### **Solution 2**

Let the sample space be:

$$
|S| = 52
$$

**a.** Drawing a Queen:

* 4 Queens in the deck (1 per suit)

$$
P(\text{Queen}) = \frac{4}{52} = \frac{1}{13}
$$

**b.** Red card or King:

* Red cards: 26 (13 Hearts + 13 Diamonds)
* Kings: 4
* Overlap: 2 red Kings (King♥, King♦)

Use inclusion-exclusion:

$$
P(\text{Red or King}) = \frac{26}{52} + \frac{4}{52} - \frac{2}{52} = \frac{28}{52} = \frac{7}{13}
$$

### **Problem 3**

> Two fair coins are tossed.

**a.** What is the probability of getting exactly one head?
**b.** What is the probability of getting at least one tail?


### **Solution 3**

Sample space:

$$
S = \{\text{HH}, \text{HT}, \text{TH}, \text{TT}\}, \quad |S| = 4
$$

**a.** Exactly one head:

* Favorable outcomes: \${\text{HT}, \text{TH}}\$

$$
P(\text{exactly one H}) = \frac{2}{4} = \frac{1}{2}
$$

**b.** At least one tail:

* Favorable outcomes: \${\text{HT}, \text{TH}, \text{TT}}\$

$$
P(\text{at least one T}) = \frac{3}{4}
$$
