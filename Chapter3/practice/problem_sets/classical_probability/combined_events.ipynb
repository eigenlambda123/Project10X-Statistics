### **Problem 1**

> Two fair six-sided dice are rolled.

**a.** What is the probability that the sum is 10?
**b.** What is the probability that the first die is 3 and the second is even?


### **Solution 1**

**Sample Space:**
Each die has 6 faces, so there are:

$$
6 \times 6 = 36 \text{ possible outcomes}
$$


**a. Probability that the sum is 10**

List all outcomes where the sum is 10:

* (4,6), (5,5), (6,4)

These are **3 favorable outcomes**, so:

$$
P(\text{sum = 10}) = \frac{3}{36} = \frac{1}{12}
$$


**b. Probability that the first die is 3 and the second is even**

Even outcomes for the second die: 2, 4, 6
So the valid outcomes are:

* (3,2), (3,4), (3,6)

These are **3 favorable outcomes**, so:

$$
P(\text{1st die = 3 and 2nd die even}) = \frac{3}{36} = \frac{1}{12}
$$

### **Problem 2**

> Two cards are drawn **without replacement** from a shuffled deck.

**a.** What is the probability that both are Aces?
**b.** What is the probability that the first is a Heart and the second is a Spade?


### **Solution 2**

Total cards = 52


**a. Probability that both are Aces**

* There are 4 Aces.
* First draw: probability of Ace = $\frac{4}{52}$
* Second draw: 3 Aces left out of 51 cards = $\frac{3}{51}$

$$
P(\text{2 Aces}) = \frac{4}{52} \times \frac{3}{51} = \frac{12}{2652} = \frac{1}{221}
$$


**b. Probability that first is a Heart, second is a Spade**

* 13 Hearts, 13 Spades

So:

$$
P(\text{Heart then Spade}) = \frac{13}{52} \times \frac{13}{51} = \frac{169}{2652} = \frac{13}{204}
$$
