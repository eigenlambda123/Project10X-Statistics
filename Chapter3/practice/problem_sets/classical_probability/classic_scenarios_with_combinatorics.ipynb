### **Problem 1**

> From a group of 5 men and 4 women, a committee of 3 people is randomly formed.

**a.** What is the probability that the committee includes **at least one woman**?
**b.** What is the probability that **all members are men**?


### **Solution 1**

Total people:

$$
5 \text{ men} + 4 \text{ women} = 9 \text{ people}
$$

Total number of 3-person committees:

$$
\binom{9}{3} = 84
$$


**a. Probability that the committee includes at least one woman**

Complement strategy:

* Number of all-male committees = choose 3 men from 5

$$
\binom{5}{3} = 10
$$

So:

$$
P(\text{at least one woman}) = 1 - \frac{10}{84} = \frac{74}{84} = \frac{37}{42}
$$


**b. Probability that all members are men**

Same as above:

$$
P(\text{all men}) = \frac{\binom{5}{3}}{\binom{9}{3}} = \frac{10}{84} = \frac{5}{42}
$$

### **Problem 2**

> Three digits are randomly selected from \${1, 2, 3, 4, 5}\$ **without replacement** to form a 3-digit number.

**a.** What is the probability that the number is **even**?
**b.** What is the probability that the number has **no repeated digits** and is **divisible by 5**?


### **Solution 2**

**Sample Space**:
All 3-digit numbers that can be formed from {1, 2, 3, 4, 5} with no repetition:

* Number of such numbers = number of **permutations of 3 out of 5**:

$$
P(5, 3) = 5 \times 4 \times 3 = 60
$$


**a. Probability that the number is even**

* An even number ends in **2 or 4** (from the set {1, 2, 3, 4, 5})

We count how many valid 3-digit numbers end with 2 or 4:

1. Fix last digit = 2:
   Remaining digits: {1, 3, 4, 5} (4 choices)
   Ways to choose first two digits = \$P(4,2) = 12\$

2. Fix last digit = 4:
   Remaining digits: {1, 2, 3, 5} â†’ again \$P(4,2) = 12\$

So favorable outcomes = \$12 + 12 = 24\$

Thus:

$$
P(\text{even}) = \frac{24}{60} = \frac{2}{5}
$$


**b. Probability that the number has no repeated digits and is divisible by 5**

A number is divisible by 5 **only** if it ends in 5 or 0. But **0 is not in the set** {1, 2, 3, 4, 5}, so we only consider numbers ending in 5.

* Fix last digit = 5
* Remaining digits = {1, 2, 3, 4}
* Ways to pick 2 digits from remaining 4 and arrange them = \$P(4, 2) = 12\$

So:

$$
P(\text{divisible by 5}) = \frac{12}{60} = \frac{1}{5}
$$
