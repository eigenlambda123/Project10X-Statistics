### **Scenario:**

We rolled a fair six-sided die 300 times. You observed the following results:

| Number | Frequency |
| ------ | --------- |
| 1      | 42        |
| 2      | 48        |
| 3      | 55        |
| 4      | 50        |
| 5      | 53        |
| 6      | 52        |


### **Goal:**

Compare **relative frequency** (empirical) vs. **theoretical probability** (which is 1/6 for each number on a fair die).


```python
import matplotlib.pyplot as plt
import pandas as pd

# Observed frequencies from the experiment
data = {
    'Number': [1, 2, 3, 4, 5, 6],
    'Frequency': [42, 48, 55, 50, 53, 52]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Total number of trials
total_rolls = sum(df['Frequency'])

# Add relative frequency and theoretical probability columns
df['Relative Frequency'] = df['Frequency'] / total_rolls
df['Theoretical Probability'] = 1/6

# Plotting
plt.figure(figsize=(10, 6))
bar_width = 0.35
x = range(len(df))

# Bars for relative frequency
plt.bar(x, df['Relative Frequency'], width=bar_width, label='Relative Frequency', color='skyblue')

# Bars for theoretical probability
plt.bar([i + bar_width for i in x], df['Theoretical Probability'], width=bar_width, label='Theoretical Probability', color='orange')

# X-axis labels
plt.xticks([i + bar_width/2 for i in x], df['Number'])

# Add labels and title
plt.xlabel('Number Rolled')
plt.ylabel('Probability')
plt.title('Relative Frequency vs. Theoretical Probability (300 Die Rolls)')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
