### **Scenario**

A school survey shows:

* 60 students play **basketball**.
* 40 students play **soccer**.
* 20 students play **both basketball and soccer**.
* The total number of students is **100**.

### **Calculations**

1. **$P(\text{Soccer} \mid \text{Basketball})$**

$$
P(\text{Soccer} \mid \text{Basketball}) = \frac{\text{Number who play both}}{\text{Number who play basketball}} 
= \frac{20}{60} 
= \frac{1}{3} \approx 0.3333
$$

2. **$P(\text{Basketball} \mid \text{Soccer})$**

$$
P(\text{Basketball} \mid \text{Soccer}) = \frac{\text{Number who play both}}{\text{Number who play soccer}} 
= \frac{20}{40} 
= \frac{1}{2} = 0.5
$$

### **Interpretation**

* **About 33%** of basketball players also play soccer.
* **Half** of soccer players also play basketball.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Data
total_students = 100
basketball = 60
soccer = 40
both = 20

# Conditional probabilities
P_soccer_given_basketball = both / basketball
P_basketball_given_soccer = both / soccer

# Create Venn diagram manually
fig, ax = plt.subplots(figsize=(6, 6))
plt.title("Conditional Probability: Students Playing Sports", fontsize=14, weight="bold")

# Draw two circles for basketball and soccer
circle1 = patches.Circle((0.4, 0.5), 0.3, fill=False, color='blue', linewidth=2, label="Basketball")
circle2 = patches.Circle((0.6, 0.5), 0.3, fill=False, color='green', linewidth=2, label="Soccer")

ax.add_patch(circle1)
ax.add_patch(circle2)

# Add numbers
plt.text(0.28, 0.5, f"{basketball - both}\nOnly BB", ha='center', va='center', fontsize=10)
plt.text(0.5, 0.5, f"{both}\nBoth", ha='center', va='center', fontsize=10, weight="bold")
plt.text(0.72, 0.5, f"{soccer - both}\nOnly SC", ha='center', va='center', fontsize=10)

# Remove axes
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Show legend
plt.legend(loc="upper right", fontsize=10)

# Display probabilities
plt.figtext(0.1, 0.2, f"P(Soccer | Basketball) = {P_soccer_given_basketball:.2f}", fontsize=11, color='blue')
plt.figtext(0.1, 0.15, f"P(Basketball | Soccer) = {P_basketball_given_soccer:.2f}", fontsize=11, color='green')

plt.show()
```
