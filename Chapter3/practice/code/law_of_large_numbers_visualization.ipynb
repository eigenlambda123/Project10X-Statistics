### **Scenario**

A fair coin is flipped many times.

* The probability of **Heads** is
  $P(\text{Heads}) = 0.5$
* The probability of **Tails** is
  $P(\text{Tails}) = 0.5$

We want to check how the **sample proportion of Heads** behaves as the number of flips increases.

### **Calculations**

1. For a **small number of flips** (say 10), the proportion of Heads might be far from 0.5 (e.g., 7/10 = 0.7).

2. For **larger flips** (say 100, 1000, 10000), the proportion of Heads gets closer to the true probability:

$$
\hat{p}_n = \frac{\text{Number of Heads in $n$ flips}}{n}
$$

and by the **Law of Large Numbers (LLN)**:

$$
\hat{p}_n \to P(\text{Heads}) = 0.5 \quad \text{as } n \to \infty
$$

### **Interpretation**

* With **few flips**, randomness causes large fluctuations.
* With **many flips**, the sample proportion stabilizes near **0.5**, the true probability.
* This illustrates the **LLN**: sample averages converge to the expected value as sample size grows.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
num_tosses = 10000  # total coin tosses
p_heads = 0.5      # true probability of heads

# Simulate coin tosses (1 = head, 0 = tail)
tosses = np.random.binomial(1, p_heads, num_tosses)

# Compute cumulative mean
cumulative_mean = np.cumsum(tosses) / np.arange(1, num_tosses + 1)

# Visualization
plt.figure(figsize=(10,6))
plt.plot(cumulative_mean, label="Sample Mean of Heads")
plt.axhline(y=p_heads, color='red', linestyle='--', label="True Probability (0.5)")
plt.xlabel("Number of Tosses")
plt.ylabel("Proportion of Heads")
plt.title("Law of Large Numbers in Action (Coin Toss Simulation)")
plt.legend()
plt.show()
```
