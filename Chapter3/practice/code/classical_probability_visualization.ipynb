### **Scenario**

We roll two fair six-sided dice. What is the probability that the **sum is 8**?

**Classical Probability Setup**

* **Sample space** = all 36 outcomes:
  $S = \{ (1,1), (1,2), \ldots, (6,6) \}$
* **Favorable outcomes** for sum = 8:
  $E = \{ (2,6), (3,5), (4,4), (5,3), (6,2) \}$

So,

$$
P(\text{sum} = 8) = \frac{|E|}{|S|} = \frac{5}{36}
$$

---

### **Visulization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample space for two dice
sample_space = [(i, j) for i in range(1, 7) for j in range(1, 7)]
event_sum_8 = [(i, j) for (i, j) in sample_space if i + j == 8]

# Plot setup
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")

# Plot all outcomes
for (i, j) in sample_space:
    if (i, j) in event_sum_8:
        plt.scatter(i, j, color='tomato', s=200, edgecolor='black', label='Sum = 8' if (i, j) == event_sum_8[0] else "")
    else:
        plt.scatter(i, j, color='lightgray', s=100, edgecolor='black', label='Other Outcomes' if (i, j) == sample_space[0] else "")

# Annotate favorable outcomes
for (i, j) in event_sum_8:
    plt.text(i, j + 0.2, f"{i}+{j}", ha='center', fontsize=9)

# Axis and labels
plt.title("Sample Space for Rolling Two Dice", fontsize=14)
plt.xlabel("Die 1", fontsize=12)
plt.ylabel("Die 2", fontsize=12)
plt.xticks(np.arange(1, 7))
plt.yticks(np.arange(1, 7))
plt.grid(True)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
